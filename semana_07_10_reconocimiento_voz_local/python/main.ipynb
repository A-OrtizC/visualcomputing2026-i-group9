```python
import speech_recognition as sr
from pythonosc.udp_client import SimpleUDPClient
import pyttsx3
```

```python
client = SimpleUDPClient("127.0.0.1", 12000)

engine = pyttsx3.init()
engine.setProperty('rate', 170)

def hablar(texto):
    engine.say(texto)
    engine.runAndWait()
```

```python
recognizer = sr.Recognizer()
mic = sr.Microphone()
```

```python
comandos_color = ["rojo", "azul", "verde"]
comandos_accion = ["girar", "iniciar", "detener"]
```

```python
while True:
    try:
        with mic as source:
            print("Escuchando...")
            recognizer.adjust_for_ambient_noise(source)
            audio = recognizer.listen(source)

        texto = recognizer.recognize_google(audio, language="es-ES")
        texto = texto.lower()

        print("Reconocido:", texto)

        if texto in comandos_color:
            client.send_message("/color", texto)
            hablar("Color " + texto)

        elif texto in comandos_accion:
            client.send_message("/accion", texto)
            hablar("Ejecutando " + texto)

        else:
            print("No válido")

    except Exception as e:
        print("Error:", e)
```
